```python
import pandas as pd
import pickle
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_pickle("train.pkl").drop("ID_code", axis=1)
test = pd.read_pickle("test.pkl")
test_id = test['ID_code'].values
test.drop(['ID_code'], axis=1, inplace=True)
```

```python
clf = RandomForestClassifier(n_estimators=100, max_depth=6,
                              random_state=0)
```

```python
X = train.drop('target', axis=1)
y = train.target
clf.fit(X, y)
```

```python
feat_import = dict(zip(X.columns, clf.feature_importances_))
feat_import = pd.DataFrame.from_dict(feat_import, orient='index')
feat_import.columns = ['importances']
feat_import = (feat_import/feat_import.max())
feat_import.sort_values(by=['importances'], ascending=False)
```
